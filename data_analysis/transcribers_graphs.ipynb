```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Users/martinasimonotti/kiparla-tools/data/transcribers_behavior.csv")
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    data=df,
    x="Trascrittore",
    y="Metalinguistic Total",
    hue="Tipo",  # differentiates From-Scratch vs Whisper-Assisted
    palette="muted"
)

plt.title("Metalinguistic Phenomena by Transcriber and Type")
plt.xlabel("Transcriber")
plt.ylabel("Count of Metalinguistic Phenomena")
plt.legend(title="Transcription Type")
plt.tight_layout()
plt.show()
```
